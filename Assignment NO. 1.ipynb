## Question 10

```python
# Function to convert an integer to binary, octal, and hexadecimal
def convert_number(number):
    # Binary conversion
    binary = ''
    for i in range(number.bit_length(), -1, -1):
        bit = (number >> i) & 1
        binary += str(bit)
    binary = binary.lstrip('0') or '0'  # Remove leading zeros or set to '0' if empty

    # Octal conversion
    octal = ''
    temp_number = number
    while temp_number > 0:
        octal = str(temp_number & 7) + octal  # Get last 3 bits (octal digit)
        temp_number >>= 3  # Shift right by 3 bits
    octal = octal or '0'  # Set to '0' if number is 0

    # Hexadecimal conversion
    hex_digits = '0123456789ABCDEF'
    hexadecimal = ''
    temp_number = number
    while temp_number > 0:
        hexadecimal = hex_digits[temp_number & 15] + hexadecimal  # Get last 4 bits (hex digit)
        temp_number >>= 4  # Shift right by 4 bits
    hexadecimal = hexadecimal or '0'  # Set to '0' if number is 0

    return binary, octal, hexadecimal

# Main program
try:
    # Taking input from the user
    number = int(input("Enter an integer: "))

    # Converting the number
    binary, octal, hexadecimal = convert_number(number)

    # Printing the results
    print(f"Binary equivalent: {binary}")
    print(f"Octal equivalent: {octal}")
    print(f"Hexadecimal equivalent: {hexadecimal}")

except ValueError:
    print("Please enter a valid integer.")
```
